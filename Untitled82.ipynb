```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("housing.csv")
data
```

```python
x = data.drop("median_house_value",axis = 1)
y = data["median_house_value"]
```

```python
x.isna().sum()
```

```python
data["ocean_proximity"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,stratify = x["ocean_proximity"])
```

```python
x_train
```

```python
x_test
```

```python
num_col = x.select_dtypes(["int","float"]).columns
num_col
```

```python
cat_col = x.select_dtypes(["object"]).columns
cat_col
```

```python
data["ocean_proximity"]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
```

```python
num_pipeline = Pipeline(
    steps = [("Impute",SimpleImputer(strategy = "median")),
            ("Scale",StandardScaler())]
)
num_pipeline
```

```python
cat_pipeline = Pipeline(
    steps = [("Encode",OneHotEncoder(sparse_output = False))]
)
cat_pipeline
```

```python
num_pipeline.fit_transform(x_train.select_dtypes(["int","float"]))
```

```python
pd.DataFrame(num_pipeline.fit_transform(x_train.select_dtypes(["int","float"])),columns = num_pipeline.get_feature_names_out())
```

```python
num_col
```

```python
print(num_col,cat_col,sep = "\n")
```

```python
from sklearn.compose import ColumnTransformer
```

```python
col_trans = ColumnTransformer(
    transformers = [("Num",num_pipeline,num_col),
                  ("Cat",cat_pipeline,cat_col)]
)
col_trans
```

```python
col_trans.fit_transform(x_train)
```

```python
pd.DataFrame(col_trans.fit_transform(x_train) , columns = col_trans.get_feature_names_out())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
final_pipeline = Pipeline(
    steps = [("Preprocess",col_trans),
            ("Model",DecisionTreeRegressor())]
)
final_pipeline
```

```python
final_pipeline.fit(x_train,y_train)
```

```python
y_pred = final_pipeline.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)*100
```
